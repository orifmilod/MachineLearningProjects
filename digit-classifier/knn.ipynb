```python
import numpy as np
import pandas as pd
import queue


class KNN:
    def __init__(self, K):
        self.K = K

    """ Input: Vector of (key, occurance_count) tuple """
    def get_most_freq(vector):
        counter = {}
        for item in vector:
            label = item[0]
            if label in counter:
                counter[label] += 1
            else:
                counter[label] = 1
        
        winner = None
        for key in counter:
            occurance = counter[key]
            if (winner ==  None):
                winner = (key, occurance)
            elif(winner[1] < occurance):
                winner = (key, occurance)

        return winner[0] 

    """Fits the training data"""
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        num_of_correct_classifcation = 0

        n = len(X_train)
        for i in range(n):
            q = queue.PriorityQueue()
            # Calculate distance to all the other data points
            for j in range(n):
                if i == j:
                    continue
                dist = self.minkowski_distance(X_train[i], X_train[j])
                q.put((dist, y_train[j]))
            
            # Get top K closes ones
            k_closest_neighbours = []
            for i in range(self.K):
                k_closest_neighbours.append(q.get())

            # Get most freq label from them
            most_freq = self.get_most_freq(k_closest_neighbours)
            if(most_freq == y_train[i]):
                num_of_correct_classifcation +=1 

        print("Correct", num_of_correct_classifcation, num_of_correct_classifcation/len(X_train))

    def minkowski_distance(self, a, b, p=2):
        distance = 0.0
        
        for i in range(len(a)):
            distance += abs(a[i] - b[i]) ** p
            
        return distance ** (1/p)

    def predict(self, X_tests, y_tests):
        num_of_correct_classifcation = 0
        q = queue.PriorityQueue()
        n = self.X_train

        for i in range(len(X_tests)):
            # Comapre it to all other training data points
            for j in range(self.X_train):
                dist = self.minkowski_distance(X_tests[i], self.X_train[j])
                q.put((dist, self.y_train[j]))
            
            # Get top K closes ones
            k_closest_neighbours = []
            for i in range(self.K):
                k_closest_neighbours.append(q.get())

            # Get most freq label from them
            most_freq = self.get_most_freq(k_closest_neighbours)
            if(most_freq == y_tests[i]):
                num_of_correct_classifcation +=1
```
